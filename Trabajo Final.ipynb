Introducción:

El parque automotor colombiano ha tenido un crecimiento considerable durante los últimos años; previo a la pandemia, fue posible notar el ingreso de múltiples marcas al mercado automotor Colombiano, junto con diferentes modelos, los cuales representan cientos de alternativas para el consumidor local.

El consumidor local con frecuencia realiza un proceso de selección del vehículo que desea adquirir según la experiencia, edad, necesidades o capacidades socioeconómicas del comprador, la variación de dichos datos es lo que determina el vehículo que será adquirido. De acuerdo con el periódico El Tiempo, los principales factores considerados por el público colombiano en general son:

1. Financiación del vehículo, es decir, cuánto se esta dispuesto a gastar en este y cómo se podrán asumir los gastos asociados a dicha compra.

2. Necesidades, esto es el rol debe cumplir el vehículo para el comprador, bien sea trabajo, transporte a la oficina, ocio, etc.

3. Soporte, que se refiere a que tan confiable es la marca que respalda esta compra.

4. Estética exterior.

5. Gadgets o aditamientos estéticos/tecnológicos en interiores.

De: https://www.eltiempo.com/archivo/documento/MAM-555827

Estos factores varían conforme a las características del individuo como: la segmentación demográfica, geográfica, la labor que realizará con el vehículo, afinidad por la estética, rasgos psicológicos, entre otros. No obstante, la mayoría de los compradores coinciden en un factor común: gastar la menor cantidad de dinero en el vehículo a cambio de obtener las mejores prestaciones, ya sea en aditamientos tecnológicos, estética o "lujitos".

Siendo así, el mercado latino ha dado una excelente respuesta a los vehículos denominados SUV (vehículo utilitario deportivo) o campero compacto; vehículos de suspensión alta preparados para las carreteras irregulares latinas, de porte familiar, bajo consumo de combustible y económicos al momento de adquirirlos.

Esta situación pone en contradicho las compañías fabricantes y les obliga a reducir costos de producción, para presentar alternativas decentes dentro del mercado latinoaméricano; por estas razones se ven en la necesidad de utilizar materiales de calidad inferior y reducir las dotaciones de seguridad de los vehículos, ya que en esencia "la seguridad" no es un factor determinante para la compra de un vehículo en el mercado latinoamericano.

Con este proyecto se pretende analizar los estándares de seguridad de algunos vehículos disponibles en el mercado colombiano y comparar los resultados obtenidos con la tasa de accidentalidad y mortalidad en la nación.

Marco Teórico:

Primeramente, los datos obtenidos del Latin NCAP brindan informacion asociada a el resultado global de la seguridad de un vehículo, convertido de 0 a 100% para facilitar su lectura en cuanto a la protección que provee un vehículo para los diferentes usuarios e información adicional de los sistemas que emplean las marcas a la hora de asistir a la seguridad.

Latin NCAP, evalúa los vehículos a partir de diferentes premisas tales como: el rendimiento biomecánico del vehículo a la hora de proteger un individuo dentro o fuera del habitáculo de impactos laterales, impactos frontales, transferencias energéticas, distribución de cargas de impacto, deformación mecánica del vehículo. Cada una de estas premisas es valorada empleando una escala de entre 0 y 5 estrellas, y considerando las posibilidades de que existan lesiones servicales, cabeza y el pecho, abdomen y la pelvis.

También se analiza la probabilidad de que ocurran lesiones en un impacto trasero, derivados del "efecto látigo", el cual puede traer lesiones considerables al cuello y la médula espinal.

Respecto al pasajero infante, se evalúan sistemas de sujeción de sillas de infante como el ISOFIX, montura en base a cinturones ya disponibles o con adición de los mismos. 

Se contrasta entonces la instalación, la protección combinada de la silla y el vehículo en pruebas dinámicas como la eyección de la silla, el contacto de la cabeza con el vehículo (en colisiones frontales, bien sea en impacto inicial o en un rebote), la excursión de la cabeza relativa al punto de inicio de la misma, por fuera de la silla. También se evalúa la fuerza transmitida al cuello y la aceleración que recibe el pecho. 

En impactos laterales, se evalúa la absorción de energía que permite la contención de la cabeza dentro del asiento y el contacto de la misma con la silla. Estos criterios entonces evalúan la probabilidad de obtener una lesión o trauma como resultado de un accidente para un infante.

También se evalúan los sistemas de desactivación automática de las bolsas de aire cuando se requiera, como en la ausencia de un pasajero.

Respecto a la protección que le ofrece a los peatones o personas desprotegidas (conductores sin cinturón en otros vehículos), esto entonces magnifica la protección que brinda el diseño de partes como la cubierta de motor, parachoques y ángulo del panorámico, así como la detección temprana de un peatón para evitar la colisión. 

Para este caso, se evalúan las áreas de impacto previamente mencionadas y cómo éstas brindan disipación de la energía frente a diferentes partes del cuerpo, como la cabeza, piernas y brazos. El momento de torsión, la suma de fuerzas y la fuerza de elongación dan como resultado del análisis la probabilidad de una lesión en la anatomía de un peatón.

En cuanto a asistencias al conductor, se pueden mencionar la presencia de sistemas como alarmas de ausencia de cinturón cuando una persona se encuentra sentada en alguna silla, bien sea trasera o delantera, asistencia a la velocidad, frenado urbano automático cuando se predice una colisión, control de estabilidad, asistencia para el cambio de carril, detección del punto ciego y la presencia de sistemas de llamada automática en caso de un accidente.

La base de datos del Latin NCAP fue creada en base a los siguientes parámetros:
- Para un automóvil que tiene una calificación de 100%, se considera que dicho vehículo brinda un 75% o más de protección al conductor, un 80% a los infantes, un 40% a los peatones y un 75% de asistencia al conductor.
- Para un automóvil con una calificación de 80%, se considera que este brinda 70% de protección al conductor, 65% a los infantes, 35% a los peatones y un 65% de asistencia al conductor.
- Para un vehículo cuya calificación es de 60%, se considera que brinda 60% de protección al conductor, 50% a los infantes, 30% a los peatones y un 50% de asistencia al conductor.
- Para un vehículo con una calificación de 40%, este brinda un 50% de protección al conductor, un 30% a infantes, un 20% a los peatones y un 40% de asistencia al conductor.
- Para un vehículo con calificación de 20%, se considera que brinda un 40% de protección al conductor, un 15% a infantes, un 10% a peatones al igual que de asistencia al conductor.
- Para un vehículo con una calificación menor a 0%, se considera un porcentaje menor a 40% de protección al conductor, menor a 15% para infantes, menor a 10% tanto para peatones como de asistencia al conductor.

Si desea conocer más información sobre los parámetros de evaluación en el Latin NCAP, se pueden conocer en: https://www.latinncap.com/es/nuestros-ensayos.

```python
#Data INIT
import pandas as pd
import csv as csv
import matplotlib.pyplot as plt
import numpy as np

#Import Raw DB from Github
urlatin = 'https://raw.githubusercontent.com/JCOM127/Data-analysis/main/LatinDB.csv'
urlsegex=  'https://raw.githubusercontent.com/JCOM127/Data-analysis/main/SegExc.csv'
urlvehiex= 'https://raw.githubusercontent.com/JCOM127/Data-analysis/main/VehiExc.csv'

latindb = pd.read_csv(urlatin, skip_blank_lines=True).to_dict()
segexcdb = pd.read_csv(urlsegex, index_col=0, skip_blank_lines=True).to_dict()
vehicexcdb = pd.read_csv(urlvehiex, index_col=0, skip_blank_lines=True).to_dict()

dflatin= pd.DataFrame(latindb)
dfsegex= pd.DataFrame(segexcdb)
dfvehicex= pd.DataFrame(vehicexcdb) 

#Data processing:

#Stat 1 - Pie Chart

dfsegex['2019'] = dfsegex['2019'].replace(',', '', regex=True) #Convert Dataframe quantity from STR to OBJ

dfsegex['2019'] = pd.to_numeric(dfsegex['2019'], errors='coerce').fillna(0).round(0).astype('Int64') #Convert DF Quantity to Float, creating and replacing NAN then converting into INT64

dfsegex['Marca'] = dfsegex.index #separate index and create new column

df_reset=dfsegex.reset_index(drop=True, level=0, inplace=True)#Reset index

dfsegex['Marca'] = dfsegex['Marca'].str.title() #Leave 'Marca' with just the Capital at first and then lowercase ;)

datasetbrand= dfsegex.groupby('Marca')['2019'].mean() #Group sets

dfsegex['Percentage'] = 100 * dfsegex['2019'] / dfsegex['2019'].sum().round(2) #Get Percentage and write it into a column

labels= dfsegex['Marca'].head(10) #Get first 10 values for labels
sizes= dfsegex['Percentage'].head(10) #Get first 10 percentages

#Graph code, remove # to activate

plt.figure(figsize=(15,15))
plt.pie(sizes, labels = labels, autopct='%.2f')
plt.title("Gráfico 1: Distribución de las ventas de vehículos nuevos en 2019 discriminado por marcas: ")
plt.show()

#Stat 2
pd.set_option("display.max_rows", None, "display.max_columns", None)

dflatin['Precio'] = dflatin['Precio'].str.replace(r'\D', '') #Remove dots from numbers
dflatin['Precio'] = pd.to_numeric(dflatin['Precio'], errors='coerce').fillna(0).astype(int) #Convert Precio into INT
glow= (dflatin['Precio']<43000000) & (dflatin['Precio']>0) #Evaluate 43M or less
glowcnt= glow.values.sum() #Return Nº of occurrences Glow True 
gmed= (dflatin['Precio']>43000000) & (dflatin['Precio']<80000000) & (dflatin['Precio']>0)#Vehicles FM 43M up to 80M
gmedcnt= gmed.values.sum() #Return Nº of occurrences Gmed True 
ghigh= (dflatin['Precio']>80000000)& (dflatin['Precio']>0) #Vehicles FM 80 UP
ghighcnt= ghigh.values.sum() #Return Nº of occurrences Ghigh True


#Vehicles set as price 0 are Not available in Colombia, 8 excluded.

vehsize= [glowcnt,gmedcnt,ghighcnt]
nomes= ['Gama Baja', 'Gama Media', 'Gama Alta']

#Graph code, remove # to activate

plt.figure(figsize=(15,15))
plt.pie(vehsize, labels = nomes, autopct='%.2f')
plt.title("Gráfico 2: Segmentos de precio en los cuales los vehículos probados por LATIN NCAP se encuentran: ")
plt.show() 


#Stat 3

dflatin['Seguridad']=dflatin['Seguridad'].astype('Int64') #Round and convert Seguridad to Int64 
latingroupdf= dflatin.groupby('Marca')['Seguridad'].mean().round(2) #Get percentage of safety across all vehicles by brand

xvals=latingroupdf.index.values.tolist() #Get index from grouped DF to obtain each brand into a list
yvals=latingroupdf.values.tolist() #Get each value for 'Seguridad' linked to Marca

#Graph code, remove # to activate

plt.figure(figsize=(15,15))
plt.bar(xvals, yvals)
plt.title("Gráfico 3: Promedio de calificación de seguridad para las marcas y sus modelos vendidos en Colombia**")#Some vehicles are not available in Colombia. Some vehicles in Colombia haven't been tested. Not all brands are available
plt.show()


#Stat 4
manufacturer= 'Renault' #Brand
modelmakesaf=dflatin.loc[dflatin['Marca'] == manufacturer] #Filter by brand
listmodels=modelmakesaf.Modelo.tolist() #List of models
segvals=modelmakesaf.Seguridad.tolist() #Safety ratings of such models

#Graph code, remove # to activate

plt.figure(figsize=(15,15))
plt.bar(listmodels, segvals)
plt.title("Gráfico 4: Modelos de la marca "+ manufacturer + ' y su promedio de calificación de seguridad')
plt.show()


#Stat 5

#Chart based on previously set manufacturer

makesegpr= dflatin.loc[dflatin['Marca'] == manufacturer] #Obtain info by brand
lcheck= (makesegpr['Precio']<43000000) & (makesegpr['Precio']>0) #Evaluate 43M or less
lcnt= lcheck.values.sum() #Return Nº of occurrences lcheck True 
mcheck= (makesegpr['Precio']>43000000) & (makesegpr['Precio']<80000000) & (makesegpr['Precio']>0)#Vehicles FM 43M up to 80M
mcnt= mcheck.values.sum() #Return Nº of occurrences mcheck True 
hcheck= (makesegpr['Precio']>80000000)& (makesegpr['Precio']>0) #Vehicles FM 80 UP
hcnt= hcheck.values.sum() #Return Nº of occurrences Ghigh True

collage=[lcnt,mcnt,hcnt] #Amount of each brand/segment hit


#Graph code, remove # to activate

plt.figure(figsize=(15,15))
plt.pie(collage, labels=nomes, autopct='%1.1f%%', pctdistance=0.85)#Plottin' info, names defined earlier.
centre_circle = plt.Circle((0, 0), 0.70, fc='white')#Create the circle for the donut
fig = plt.gcf()
fig.gca().add_artist(centre_circle)# Adding Circle in Pie chart
plt.title("Gráfico 5: Marca discriminada por segmentos de precio")# Adding Title of chart
plt.show() #Show that thing


# Stat 6

totalvehicles19= dfsegex['2019'].sum() #Get all the vehicles sold in 2019.
maketocompare= dfsegex.loc[dfsegex.Marca == manufacturer] #Get Manufacturer previously set info.
finalamount= maketocompare['2019'].item() #Amount of vehicles sold per manufacturer previously set. Item gets the value and not the object
makesafe= makesegpr['Seguridad'].mean().round(2) #AVG Safety rating per manufacturer


#Graph code, remove # to activate

ylabelf=[totalvehicles19,finalamount] #Labels
xvalf=['Total 2019', manufacturer]#Labels

# Placing the plots in the plane
plot1 = plt.subplot2grid((1, 3), (0, 0), colspan=1)
plot2 = plt.subplot2grid((3, 4), (0, 2), rowspan=3, colspan=2)


#Plot for Vehicles sold/Manufacturer
plot2.bar(xvalf, ylabelf,width=0.5,)
plot2.set_title("6.1: Vehiculos Vendidos/ Vehículos por marca")
  
# Plot for AVG Safety rating
plot1.bar(manufacturer, [makesafe],width=0.5)
plot1.set_title('6.2: Promedio de calificación de seguridad' + '\n' ' de la marca seleccionada' ,loc='right')
plot1.set_yticks(np.arange(0, 110, step=10)) #Set Y axis for range 0-100 step 10
plot1.set_ylim(0,100) #Y Axis max value is 100

# Packing all the plots and displaying them
plt.figure(figsize=(12, 12))
plt.tight_layout()
plt.show()


#Stat 7

dfvehicex['2019'] = dfvehicex['2019'].str.replace(r'\D', '') #remove dots from numbers
dfvehicex['2019'] = pd.to_numeric(dfvehicex['2019'], errors='coerce').fillna(0).astype(int) #Convert to int
dfvehicex['Vehiculo'] = dfvehicex.index #Get a column after every vehicle from index
indexreset=dfvehicex.reset_index(drop=True, level=0, inplace=True)#Reset index
dfvehicex['Vehiculo'] = dfvehicex['Vehiculo'].str.title() #Each word with Capital at first.

vehicles19= dfvehicex['Vehiculo'].head(10) #Get first 10 
size19= dfsegex['2019'].head(10) #Get first 10 

#Graph code, remove # to activate

plt.figure(figsize=(15,15))
plt.pie(size19, labels = vehicles19, autopct='%.2f')
plt.title("Gráfico 7: Vehículos más vendidos en Colombia en 2019: ")
plt.show()

vehiclelookup= "Sandero" #Introduce Vehicle name or partial name to be looked up:

filter1 = dfvehicex[dfvehicex.Vehiculo.str.contains(vehiclelookup,case=False)]
filter2 = dflatin[dflatin.Modelo.str.contains(vehiclelookup,case=False)]

display(filter1) #Shows list of partial results that contains the Vehicle name set previously. Units sold
display(filter2) #Shows list of partial results that contains the Vehicle name set previously. Safety
```
